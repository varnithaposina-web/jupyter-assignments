```python
# EMPLOYEE PERFORMANCE BONUS ELIGIBILITY PROBLEM:

employees = {"Ravi": 92, "Anita": 88, "Kiran": 92, "Suresh": 85} 

# Find the highest score in the dictionary
max_score = max(employees.values())

#Use list comprehension to find all names that match the max score
top_performers = [name for name, score in employees.items() if score == max_score]

# Display results formatted as a comma-separated string
print(f"Top Performers Eligible for Bonus: {', ' .join(top_performers)} (Score: {max_score})")
```

```python
# SEARCH QUERY KEYWORD ANALYSIS PROBLEM:

import re
query = "Buy mobile phone buy phone online"

# Fix: We lowercase and remove punctuation FIRST.
# Then we call .split() on the resulting string.
clean_query = re.sub(r'[^\w\s]', '', query.lower())
words = clean_query.split()

# Count frequency only if the word appears more than once
freq = {w: words.count(w) for w in set(words) if words.count(w) > 1}

print(freq)
```

```python
# SENSOR DATA VALIDATION PROBLEM:

sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# Use enumerate to get both index (hour) and value
# val % 2 == 0 checks if the number is even
valid_readings = [(i, val) for i, val in enumerate(sensor_readings) if val % 2 == 0]

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
# EMAIL DOMAIN USAGE ANALYSIS PROBLEM:

emails = ["ravi@gmail.com", "anita@yahoo.com", "kiran@gmail.com", "suresh@gmail.com", "meena@gmail.com"]

# Extract the domain part (after '@') for every email
domains = [email.split('@')[1] for email in emails]

# Create a unique set of domains and count their occurences
counts = {d: domains.count(d) for d in set(domains)}

# Calculate and display percentage for each domain
for domain, count in counts.items():
    percentage = (count / len(emails)) * 100
    print(f"{domain}: {percentage:.0f}%")

```

```python
# DUPLICATE USER ID DETECTION PROBLEM:

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Identify IDs that appear more than once using a dictionary comprehension
duplicates = {uid: user_ids.count(uid) for uid in set(user_ids) if user_ids.count(uid) > 1}

# Print the formatted output for each duplicate found
for uid, count in duplicates.items():
    print(f"{uid} -> {count} times")
```
