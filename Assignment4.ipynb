```python
# MOBILE RECHARGE VALIDATION SYSTEM PROBLEM:


# Function to validate mobile recharge
def validate_recharge():
    # List of valid recharge plans
    valid_plans = [199, 299, 399, 599]

    # Retry until valid recharge is entered
    while True:
        amount = int(input("Enter recharge amount: "))

        # Check minimum amount and valid plan
        if amount >= 50 and amount in valid_plans:
            print("Recharge Successful")
            break
        else:
            print("Invalid recharge. Please try again.")

# Function call
validate_recharge()
```

```python
# INVENTORY REORDER ALERT SYSTEM PROBLEM:

# Function to check stock levels
def reorder_alert(stock):
    # Loop through dictionary items
    for product, quantity in stock.items():
        # Check if stock is below minimum limit
        if quantity < 15:
            print(product, ": Reorder Alert")
        else:
            print(product, ": Stock OK")

# Dictionary of products and quantities
products = {"Rice": 10, "Oil": 25, "Sugar": 8}

# Function call
reorder_alert(products)
```

```python
# STUDENT RESULT PROCESSING SYSTEM PROBLEM:

# Function to calculate result
def student_result(marks):
    total = 0

    # Calculate total marks
    for m in marks:
        total += m

    # Calculate average
    average = total / len(marks)

    # Return result based on average
    if average >= 50:
        return "Pass"
    else:
        return "Fail"

# Function call
print(student_result([60, 55, 40, 50]))
```

```python
# CAB FARE ESTIMATOR WITH RETRY OPTION

# Function to calculate cab fare
def calculate_fare(km, peak):
    # Base fare + distance charge
    fare = 50 + (km * 12)

    # Add 25% extra during peak hours
    if peak:
        fare = fare + (fare * 0.25)

    return fare

# Retry loop
while True:
    km = int(input("Enter distance in km: "))
    peak = input("Is it peak hour? (y/n): ") == 'y'

    # Display fare
    print("Total Fare:", calculate_fare(km, peak))

    # Ask user to retry
    if input("Retry? (y/n): ") != 'y':
        break
```

```python
# EMPLOYEE ATTENDEANCE ELIGIBILITY CHECKER PROBLEM:

# Function to check attendance eligibility
def attendance_check(attendance):
    present_days = 0

    # Count present days
    for day in attendance:
        if day == "P":
            present_days += 1

    # Calculate attendance percentage
    percentage = (present_days / len(attendance)) * 100

    # Check eligibility
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

# Function call
print(attendance_check(["P", "P", "A", "P", "P"]))
```

```python
# PASSWORD STRENGTH CHECKER PROBLEM:

# Function to check password strength
def check_password(password):
    has_digit = False
    has_special = False

    # Loop through each character in password
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in "@#$":
            has_special = True

    # Validate all conditions
    if len(password) >= 8 and has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"

# Function call
print(check_password("Pass@123"))


```
