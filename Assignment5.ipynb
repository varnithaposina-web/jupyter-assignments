```python
# SOCIAL MEDIA - POST ENGAGEMENT ANALYZER PROBLEM:

def analyze_engagement(likes_list):
    # Calculate total likes using a loop
    total_likes = 0
    for like in likes_list:
        total_likes += like
    
    # Determine post status
    status = "Viral Post" if total_likes >= 1000 else "Normal Engagement"
    
    print(f"Total Likes: {total_likes}")
    print(f"Post Status: {status}")

# Input: List of likes from different sources/timeframes
analyze_engagement([400, 300, 350])
```

```python
# HEALTHCARE - MEDICINE STOCK ALERT SYSTEM PROBLEM: 

def check_medicine_stock(stock_level):
    # Use conditional logic to check stock
    if stock_level < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"
    
    print(f"Medicine Stock: {stock_level}")
    print(f"Status: {status}")

# Input: Current stock count
check_medicine_stock(6)
```

```python
# AGRICULTURE - RAINFALL ADEQUACY CHECKER PROBLEM:

def check_rainfall(rainfall_data, required_level):
    # Calculate average using sum and length
    total_rainfall = 0
    for rain in rainfall_data:
        total_rainfall += rain
    
    average_rainfall = total_rainfall / len(rainfall_data)
    
    # Check adequacy
    status = "Adequate Rainfall" if average_rainfall >= required_level else "Inadequate Rainfall"
    
    print(f"Average Rainfall: {average_rainfall:.0f}")
    print(f"Rainfall Status: {status}")

# Input: Weekly rainfall readings and required average
check_rainfall([80, 60, 70, 78], 70)
```

```python
# SOCIAL MEDIA - DUPLICATE ACCOUNT DETECTION PROBLEM:

def detect_duplicates(usernames):
    # A set only stores unique values; if lengths differ, duplicates exist
    if len(usernames) != len(set(usernames)):
        result = "Yes"
    else:
        result = "No"
    
    print(f"Duplicate Accounts Found: {result}")

# Input: List of registered usernames
detect_duplicates(["user1", "ace_99", "user1", "dev_pro"])
```

```python
# HEALTHCARE - APPOINTMENT ELIGIBILITY CHECKER PROBLEM:

def check_appointment_eligibility(age):
    # Check if patient is an adult
    if age >= 18:
        status = "Eligible"
    else:
        status = "Not Eligible"
    
    print(f"Patient Age: {age}")
    print(f"Eligibility Status: {status}")

# Input: Patient's age
check_appointment_eligibility(21)
```

```python
# AGRICULTURE - PREMIUM CROP PRICE FILTER PROBLEM:

def filter_premium_crops(price_list):
    # Use list comprehension to filter prices above 2000
    premium_crops = [price for price in price_list if price > 2000]
    
    print(f"Premium Crops: {premium_crops}")

# Input: Market prices of various crops
filter_premium_crops([1500, 2500, 1800, 3200, 900])
```

```python
# SYSTEM MONITORING - APPLICATION HEALTH CHECKER PROBLEM:

def check_app_health(error_count):
    # Multiple conditions for system health
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    
    print(f"Error Count: {error_count}")
    print(f"System Status: {status}")

# Input: Detected system errors
check_app_health(7)
```

```python
# BANKING - DAILY TRANSACTION LIMIT CHECKER PROBLEM:

def validate_transaction(amount):
    daily_limit = 50000
    
    # Check if amount is within the bank's safety limit
    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"
    
    print(f"Transaction Amount: {amount}")
    print(f"Transaction Status: {status}")

# Input: Attempted transaction amount
validate_transaction(60000)
```

```python
# E-LEARNING - STUDENT ATTENDANCE ELIGIBILITY SYSTEM PROBLEM:

def check_exam_eligibility(attendance_list):
    # Calculate percentage: (Days Present / Total Days) * 100
    total_days = len(attendance_list)
    days_present = sum(attendance_list) # Assuming 1 for present, 0 for absent
    
    percentage = (days_present / total_days) * 100
    
    status = "Eligible" if percentage >= 75 else "Not Eligible"
    
    print(f"Attendance Percentage: {percentage:.1f}")
    print(f"Exam Eligibility: {status}")

# Input: Binary list (1 = present, 0 = absent)
check_exam_eligibility([1, 1, 1, 0, 1])
```

```python
# SMART ELECTRICITY BILL ANALYZER PROBLEM:

def calculate_electricity_bill(units):
    bill = 0
    
    # Step 1: Calculate Bill based on Slabs
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    
    # Step 2: Classify Usage
    if bill < 500:
        usage = "Low Usage"
    elif 500 <= bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"
        
    return bill, usage

# Input: Units consumed
units_consumed = 250
total_bill, usage_status = calculate_electricity_bill(units_consumed)

print(f"Units Consumed: {units_consumed}")
print(f"Total Bill: ₹{total_bill}")
print(f"Usage Classification: {usage_status}")
```
