```python
#UNIQUE WORDS IN A SENTENCE
#Input sentence
sentence = "python is easy and python is powerful"

#Step 1: Split the sentence into a list of words
words = sentence.split()

#Step 2: Convert the list to a set to remove duplicates
unique_words = set(words)

#Step 3: Display results
print(f"Unique words count: {len(unique_words)}")
print(f"Unique words: {unique_words}")
```

```python
#HIGHEST SALARY FROM EMPLOYEE DATA
employees = {
    "Ravi": 75000,
    "Anita":68000,
    "Kiran": 72000
}

# Use max() on the dictionary keys, comparing them by their dictionary values
highest_paid_employee = max(employees, key=employees.get)
highest_salary = employees[highest_paid_employee]

print(f"Highest Salary: {highest_paid_employee} - {highest_salary}")
```

```python
#FIND MAXIMUM AND MINIMUM VALUES
numbers = [45, 22, 89, 10,66]

# Initialize max and min with the first element
maximum = numbers[0]
minimum = numbers[0]

# Loop through the list to compare values
for num in numbers:
    if num>maximum:
        maximum = num
    if num < minimum:
        minimum = num

print(f"List: {numbers}")
print(f"Max: {maximum}")
print(f"Min: {minimum}")
```

```python
#COUNT PRODUCTS ABOVE A PRICE THRESHOLD
prices = [450, 1200, 899, 1500, 300]
threshold = 1000
count = 0

# Check each price against the threshold
for price in prices:
    if price > threshold:
        count+=1

print(f"Products above 1000: {count}")
```

```python
#CALCULATE ATTENDANCE PERCENTAGE
attendance = ["P", "P", "A", "P", "P"]

# Count "P" for Present
present_count = attendance.count("P")
total_days= len(attendance)

#Formula: (Part / Total) * 100
percentage = (present_count / total_days) * 100

print(f"Attendence Percentage: {percentage}")
```

```python
#REMOVE DUPLICATE PHONE NUMBERS
phone_numbers = [9876543210, 9123456789, 9876543210]

# Converting to a set removes duplicates automatically
unique_phones = set(phone_numbers)

print(f"Unique phone numbers: {unique_phones}")
```

```python
#COUNT CHARACTER FREQUENCY
# 1.Define the input string
text = "pythonp"
frequency ={}

# Iterate through each character
for char in text:
    # If char is already in dict, increment; otherwise, set to 1
    frequency[char] = frequency.get(char, 0)+1

print(frequency)
```
